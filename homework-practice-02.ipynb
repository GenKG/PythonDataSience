```python
# Задача 1
x = np.array([[1, 0, 1], 
              [2, 0, 2], 
              [3, 0, 3], 
              [4, 4, 4]])
def vector1(x):
    d = np.diag(x)
    z = d[d.nonzero()]
    return np.prod(z)
vector1(x)
```

```python
def nonvector(a):
    prod = 1
    for i in range(a.shape[0]):
        for j in range(a.shape[1]):
        
            if i==j and a[i,j]!= 0:
                prod = prod * a[i, j]
    return prod
nonvector(x)
```

```python
def nonvect_1(a):
    prod = 1
    for i in range(len(a)):
        try:
            if a[i,i]!= 0:
                prod = prod * a[i, i]
        except:
            pass
    return prod
```

```python
import time

def get_array(arr_size=10, dim = 2):
    if dim == 2:
        arr = np.random.rand(arr_size, arr_size)*100
    elif dim ==1:
        arr = np.random.rand(arr_size)*100
    elif dim ==3:
        arr = np.random.rand(arr_size,arr_size,3)*100
    return arr

def get_time(
        vect,
        nonvect,
        sizes=(10, 100, 200, 400, 500, 600, 800,
                     1000, 2000, 4000, 6000, 8000),
        dim = 2,
        num_arrays = 1,
        verbose=False):

    t_vect = []
    t_nvect = []
    for arr_size in sizes:
        if verbose:
            print(f'computing for {arr_size}x{arr_size} array')
        arrs = []
        for i in range(num_arrays):
            arrs.append(get_array(arr_size, dim))
        
            
        t_start = time.perf_counter()
        vect(*arrs)
        t_end = time.perf_counter()
        t_vect.append(t_end-t_start)

        t_start = time.perf_counter()
        nonvect(*arrs)
        t_end = time.perf_counter()
        t_nvect.append(t_end-t_start)

    plt.figure(figsize=(12, 8))
    plt.title('Сравнение времени выполнения функций')
    plt.plot(sizes, np.array(t_vect)*100, label='Векторизованная')
    plt.plot(sizes, np.array(t_nvect)*100, label='Невекторизованная')
    plt.xlabel(u"Длина массива")
    plt.ylabel(u"Время (мс)")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def pairplot(df, target):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * 4, nrow * 4))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature)
        if i % ncol == 0:
            plt.ylabel('target')
```
